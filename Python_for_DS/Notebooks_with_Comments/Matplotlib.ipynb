 # **MatPlotLib Tutorial**

 matplotlib is a plotting library for the python programming language and its numerical mathematics extension Numpy . it provides an object-oriented API for embedding plots into application using genral-purpose GUI toolkits like Tkinter , wxpython , Qt or GTK+ . 

some of the major pros of matplotlib are :

- Generally easy to get started for simple plots
- Support for custom labels and text
- great control of every elememnt in a figure
- high-quality ouput in many formats
- very customizable in genral

```python
## plotting using matplotlib 
## plt scatter  # use to sccatter the values of x and y in 2d plot 

plt.scatter(x,y,c='g')
plt.xlabel('X Axis')
plt.ylabel('Y Axis')
plt.title('Graph in 2D') # not showing graph right becauese for showing graph last line need to be plotting command 
plt.savefig('graph.png')
plt.show()
```

```python
plt.plot(x,y ,'ro--'  )
plt.show()
```

```python
plt.plot(x,y ,'r' , linestyle='dashed' ,linewidth=2 , markersize=12)
plt.xlabel('X axis')
plt.ylabel('Y axis')
plt.title('2d diagram ')
plt.show()
```

```python
## creating subplots
plt.subplot(2,2,1) # here we defined that we are using 2 rows and 2 coulumns 2,2 means we just have 4 positiions .
plt.plot(x,y,'g*--')
plt.subplot(2,2,2)
plt.plot(x,y,'ro--')
plt.subplot(2,2,3)
plt.plot(x,y,'o--')
plt.subplot(2,2,4)
plt.plot(x,y,'bo')
plt.show()
```

```python
# compute the x and y coordinates for points on a sine curve  . 
x= np.arange(0,4*np.pi,0.1)
y = np.sin(x)
plt.title("sine wave form ")

# plot the points using matplotlib 
plt.plot(x,y)
plt.show()
```

```python
# subplot 
# compute the x and y coordinates for points on sine and cosine curves
x = np.arange(0,5*np.pi,0.1)
y_sin = np.sin(x)
y_cos = np.cos(x)

# set up a subplot grid that has height 2 and width 1 , 
# and set the first such subplot as active .
plt.subplot(2,1,1)

#make the first subplot
plt.plot(x,y_sin , 'g--')
plt.title('Sine')

# set the 2nd subplot as active , and make the second plot .
plt.subplot(2,1,2)
plt.plot(x,y_cos , 'r--')
plt.title('cosine')

# show the figure 
plt.show()
```

```python
# bar plot 
x=[2,8,10]
y=[11,16,9]

x2 = [3,9,11]
y2 = [6,15,7]
plt.bar(x,y)
plt.bar(x2,y2,color = 'g')
plt.title('Bar Graph')
plt.ylabel('Y axis')
plt.xlabel('X axis')

plt.show()
```

# **Box Plot Using Matplotlib**

```python
data = [np.random.normal(0, std,100) for std in range(1,4)] # three list of values created and each list will be having one different array  

# rectangular box plot , sir discussed this in stastics also 

plt.boxplot(data , vert=True,patch_artist=True);
plt.show()
```

# PIE CHART 

```python
# Data to plot 
labels = 'Python' , 'C++' , 'Julia' , 'R'
sizes = [215 , 153 , 253 , 244]
colors = ['gold' , 'yellowgreen' , 'lightcoral' , 'lightskyblue']
explode = (0.1 , 0 , 0 , 0) # explode 1st slice 

# plot 

plt.pie(sizes , explode=explode , labels=labels ,  colors=colors ,
autopct='%1.1f%%' , shadow=True)

plt.axis('equal')
plt.show()
```
