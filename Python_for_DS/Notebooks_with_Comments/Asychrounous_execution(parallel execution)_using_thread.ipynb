```python
import time 
from concurrent.futures import ThreadPoolExecutor
# what this function is doing whatever parameter i am giving it , it is just returning that same value . 
def returnnumber(a):
    return a 
```

```python
returnnumber(1)
```

```python
# now here , we are going to find out how much it is taking by thread pool executor to execute this process using the time
start = time.time()
with ThreadPoolExecutor(max_workers=None) as executor:
    for result in executor.map(returnnumber,range(100)):
        print('count: {0}'.format(result))
# after process completion we are going to print out the time here by taking out the difference of starting time and after the execution time 

print('the total time is : {0}'.format(time.time()-start))
```

```python
# now what we are doing here is we are adding the time.sleep here to pause a 1 sec for each execution of the function.

# if we wont use the time.sleep then that code process of counting will get executed in just a blink of eye because of our laptop processing power . even if we use just 1 worker i.e  one thread . 

import time 
from concurrent.futures import ThreadPoolExecutor
# what this function is doing whatever parameter i am giving it , it is just returning that same value . 
def returnnumber(a):
    time.sleep(1)
    return a 
```

```python
# now we are going to the change the max_workers to 5 , that basically means that as soon as i start it 5 thread wiill be created and all those 5 threads calling this particular number parallely

# cloned the code 

# what is actully happening is each thread having a number and all are working or processing it parallely and returning the number simulatenously , this like for big data this is very useful to distribute our work .
# like whenever output for like 5 workers after each 1 sec sleep there is 5 output is coming because the workers working paralley and returning the number simulataneously . 

# now here , we are going to find out how much it is taking by thread pool executor to execute this process using the time

start = time.time()
with ThreadPoolExecutor(max_workers=5) as executor:
    for result in executor.map(returnnumber,range(100)):
        print('count: {0}'.format(result))
# after process completion we are going to print out the time here by taking out the difference of starting time and after the execution time 

print('the total time is : {0}'.format(time.time()-start))
```
