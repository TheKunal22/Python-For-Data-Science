```python
data = ('index,a,b,c\n'
       '4,apple,bat,4.7\n'
       '8,mango,cow,10\n')
```

```python
pd.read_csv(StringIO(data)) 
```

```python
pd.read_csv(StringIO(data),index_col=0)
# here we are just making cloumns to our index cloumns by defulat 0 and 1 h but we can also set our index columns too 
# index column matlb vahi h bhai 0 , 1 ,2 ,3 .... indexs like in arrays 
# toh apn apne hisab s ye decide kar sakte h kisko apn ko index column banana h 
# abh yaha p 4 and 8 ko ye index column lele raha h toh agar 4.7 and 10.0 nahi hote toh vaha toh NAN  a jata toh apn toh 4 and 8 ko vo indxing na mane uske liye we willl use index_col=False
```

```python
pd.read_csv(StringIO(data),index_col=1) # yaha p column 1 index column ban gya simple h bohot 
```

```python
pd.read_csv(StringIO(data),index_col=False) # now as you can see now 4 and 8 isnt index anymore it is taking 0 and 1 as a index after doing index_cols = false varna ye 1 column ko toh by deafult index  hi maan raha tha 
```
